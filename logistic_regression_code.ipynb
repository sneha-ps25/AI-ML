```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv('Social_Network_Ads.csv')
x= dataset.iloc[:,:-1].values
y= dataset.iloc[:,-1].values
```

```python
dataset
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print(x_train)
```

```python
print(y_train)
```

```python
print(x_test)
```

```python
print(y_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train= sc.fit_transform(x_train)
x_test= sc.transform(x_test)
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(x_train,y_train)
```

```python
print(classifier.predict(sc.transform([[50,40000]])))
```

```python
y_pred=classifier.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,y_pred))
```

```python
from sklearn import metrics
y_pred_train=classifier.predict(x_train)
print(metrics.accuracy_score(y_train,y_pred_train))
print(classification_report(y_train,y_pred_train))
```

```python
from sklearn.inspection import DecisionBoundaryDisplay
DecisionBoundaryDisplay.from_estimator(
    classifier,x_train,response_method="predict",cmap=plt.cm.Spectral,alpha=0.8,xlabel=dataset.columns[0],ylabel=dataset.columns[1])
plt.scatter(x_train[:,0],x_train[:,1],c=y_train,s=20,edgecolors="k")
plt.title('Logistic Regression(Training set)')
plt.show()
```

```python
from sklearn.inspection import DecisionBoundaryDisplay
DecisionBoundaryDisplay.from_estimator(
    classifier,x_test,response_method="predict",cmap=plt.cm.Spectral,alpha=0.8,xlabel=dataset.columns[0],ylabel=dataset.columns[1])
plt.scatter(x_test[:,0],x_test[:,1],c=y_test,s=20,edgecolors="k")
plt.title('Logistic Regression (Testing set)')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
(x_cancer,y_cancer)= load_breast_cancer(return_X_y=True)
x_train,x_test,y_train,y_test=train_test_split(x_cancer,y_cancer)
clf=LogisticRegression(max_iter=10000).fit(x_train,y_train)
clf.score(x_train,y_train)
clf.score(x_test,y_test)
```

```python
import joblib

# Save the model
joblib.dump(clf, "model.joblib")

# Load the model later
loaded_model = joblib.load("model.joblib")
```

```python
import pickle

# Save
with open("model.pkl", "wb") as file:
    pickle.dump(clf, file)

# Load
with open("model.pkl", "rb") as file:
    loaded_model = pickle.load(file)
```
